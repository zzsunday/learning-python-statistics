## （2）数据操作

（1）读取香港酒店数据。

```python
df=pd.read_excel(r'C:\Users\Administrator\Desktop\python数据分析作业\香港酒店数据.xlsx')
df.head()
```

（2）按照数据的内容，重新设置数据的索引，重新设置列名称为'名字','类型','城市','地区','地点','评分','评分人数','价格'。

```python
df.columns = ['名字','类型','城市','地区','地点','评分','评分人数','价格']
df
```

（3）查看所有类型为“浪漫情侣”的酒店

```python
df[df['类型']== '浪漫情侣']
```

（4）查看所有类型为“浪漫情侣”，地区在湾仔的酒店

```python
df[(df['类型']== '浪漫情侣')&(df['地区']=='湾仔')]
```

（5）查看所有地址在观塘或者油尖旺，评分大于4的酒店

```python
df[(df['地区']== '观塘')|(df['地区']=='油尖旺')&(df['评分']>4)]
```

（6）查看类型缺失的数据

```python
df[df['类型'].isnull()]
```

（7）用“其他”填充类型和地区

```python
df['类型'].fillna('其他', inplace=True)
df['地区'].fillna('其他',inplace=True)
```

（8）用评分均值填充缺失值

```python
df['评分'].fillna(np.mean(df['评分']),inplace = True)
```

（9）删除价格和评分人数的缺失值

```python
df[df['价格'].isnull()]
```

```python
df
```

```python
df.dropna(subset=['价格'],inplace=True)
df.dropna(subset=['评分人数'],inplace=True)
```
