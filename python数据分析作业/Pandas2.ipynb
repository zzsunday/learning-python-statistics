### 2.4 数据透视 


#### 使用pandas中的pivot_table的核心是你能很好的理解自己的数据，并轻清楚的知道你想通过透视表解决什么问题。

#### 基础形式

```python
pd.set_option('max_columns',50)
pd.set_option('max_rows',50)  #最大的展现行数，50行之前不会出现省略号
```

```python
pd.pivot_table(df,index=['年代']) #以年代为索引，对投票人数，时长，评分来进行均值的计算
```

#### 也可以多个索引

```python
pd.pivot_table(df,index=['年代','产地'])
```

#### 也可以指定需要统计汇总的数据

```python
pd.pivot_table(df,index=['年代','产地'],values=['评分'])
```

#### 还可以指定函数，来统计不同的统计值

pd.pivot_table(df,index=['年代','产地'],values=['投票人数'],aggfunc=np.sum)

#### 对"产地"实现数据聚合，通过“投票”和”评分“进行对应分组总结

```python
pd.pivot_table(df,index=['产地'],values=['投票人数','评分'],aggfunc=[np.sum,np.mean])
```

#### 非数值（NAN）可以使用fill_value将其设置为0

```python
pd.pivot_table(df,index=['年代'],aggfunc=[np.sum,np.mean],fill_value=0)
```

#### margins=True,可以在下方显示一些总和的数据

```python
pd.pivot_table(df,index=['年代'],aggfunc=[np.sum,np.mean],fill_value=0,margins=True)
```

#### 对不同值执行不同的函数,可以向aggfunc传递一个字典

```python
pd.pivot_table(df,index=['产地'],values=['投票人数','评分'],aggfunc={'投票人数':np.sum,'评分':np.mean})
```

```python
pd.pivot_table(df,index=['年代'],values=['投票人数','评分'],aggfunc={'投票人数':np.sum,'评分':np.mean},fill_value=0,margins=True)
```

#### 透视表过滤

```python
table=pd.pivot_table(df,index=['年代'],values=['投票人数','评分'],aggfunc={'投票人数':np.sum,'评分':np.mean})
```
