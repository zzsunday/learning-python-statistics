### \d :数字字符，只匹配一次   \D:非数字字符，特殊字符

```python
print(re.findall('\d',text))
```

```python
print(re.findall('\D',text))
```

###  +:匹配一个或多个表达式

```python
print(re.findall('\d+',text))
```

```python
print(re.findall(r'[1-9]\d{5}',text)) 
```

```python
print(re.findall(r'\d\d\d\d\d\d+',text)) #匹配大于等于6位数以上
```

###  \*：匹配零个或多个表达式 

```python
print(re.findall(r'\d\d\d\d\d*',text))#匹配大于等于4
```

### .:匹配任意字符，除了换行符，当re.DOTALL标记被指定时，则可以匹配包括换行符的任意字符。 

```python
print(re.findall(r'\d\d\d\d\d\d+.',text)) 
```

```python
print(re.findall(r'\d\d\d\d\d\d+..',text)) #
```

```python
print(re.findall(r'\d\d\d\d\d\d+.{3}',text))
```

###  .\*:匹配任意字符

```python
print(re.findall(r'\d\d\d\d\d\d+.*',text))
```

```python
print(re.findall(r'\d\d\d\d\d.*@',text))
```

```python
print(re.findall(r'[a-z0-9][a-z0-9][a-z0-9][a-z0-9][a-z0-9].*@',text))
```

```python
print(re.findall(r'\d\d\d\d\d\d*',text))
```

```python
print(re.findall(r'\d\d\d\d\d\d*.',text))
```

###  ？：	匹配0个或1个由前面的正则表达式定义的片段，非贪婪方式（有零个就匹配零个，有一个就匹配一个） 

```python
print(re.findall(r'\d\d\d\d\d\d*?',text))
```

```python
print(re.findall(r'\d\d\d\d\d\d*',text))
```

```python
print(re.findall(r'\d\d\d\d\d\d*.*',text))
```

```python
print(re.findall(r'\d\d\d\d\d\d*.*?',text))
```

### re{ n, m}:匹配 n 到 m 次由前面的正则表达式定义的片段，贪婪方式

```python
print(re.findall(r'[A-Za-z0-9]{5,15}',text))
```

### 泛匹配和精确匹配 ：（）


```python
print(re.findall(r'[A-z0-9]{5,15}@[A-z0-9]{1,5}.[A-z0-9]{1,5}',text))
```

```python
print(re.findall(r'([A-z0-9]{5,15})@([A-z0-9]{1,5}).([A-z0-9]{1,5})',text)) #注意括号的使用在哪
```

```python
print(re.findall('.*?}',text))
```
